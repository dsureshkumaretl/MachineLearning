```r
histogram <- function(x)
{
    title <- paste(deparse(substitute(x), 500), collapse="\n")
    sdx <- sd(x)
    mx <- mean(x)
    hist(x, prob=TRUE,
         main=paste("Histogram of ",title),
         xlim=c(mx-3*sdx, mx+3*sdx), ylim=c(0, 0.5/sdx))
    curve(dnorm(x, mean=mx, sd=sdx), col='orange', lwd=3, add=TRUE)
}

par(mfrow=c(2,2))
histogram(housedata_vect$price)
histogram(log(housedata_vect$price))
histogram(housedata_vect$bedrooms)
histogram(log(housedata_vect$bedrooms +1))
histogram(housedata_vect$bathrooms)
histogram(log(housedata_vect$bathrooms +1))
histogram(housedata_vect$sqft_living)
histogram(log(housedata_vect$sqft_living +1))
histogram(housedata_vect$sqft_lot)
histogram(log(housedata_vect$sqft_lot +1))
histogram(housedata_vect$floors)
histogram(log(housedata_vect$floors +1))
histogram(housedata_vect$waterfront)
histogram(log(housedata_vect$waterfront +1))
histogram(housedata_vect$view)
histogram(log(housedata_vect$view +1))
histogram(housedata_vect$condition)
histogram(log(housedata_vect$condition +1))
histogram(housedata_vect$grade)
histogram(log(housedata_vect$grade +1))
histogram(housedata_vect$sqft_above)
histogram(log(housedata_vect$sqft_above +1))
histogram(housedata_vect$sqft_basement)
histogram(log(housedata_vect$sqft_basement +1))
histogram(housedata_vect$yr_built)
histogram(log(housedata_vect$yr_built +1))
histogram(housedata_vect$yr_renovated)
histogram(log(housedata_vect$yr_renovated +1))
histogram(housedata_vect$zipcode)
histogram(log(housedata_vect$zipcode +1))
histogram(housedata_vect$lat)
histogram(log(housedata_vect$lat +1))
histogram(housedata_vect$sqft_living15)
histogram(log(housedata_vect$sqft_living15 +1))
histogram(housedata_vect$sqft_lot15)
histogram(log(housedata_vect$sqft_lot15 +1))
histogram(housedata_vect$long)
histogram(log(housedata_vect$long + 1))
```

```r
cat("The above result from the box plot suggest that above the box we see a significant number of samples are exptrapolated as an outliers and hence not in the actuals. It is very important that we excercise care in handling  these outliers (extremities) , making changes to these data which are genuinely observed values is quite challenging .\nIt has a considerable impact in the predictible model , in order to understand the effect let us consider the linear regression both inlcuding and excluding the outliers - For the same scenario let us do the following ,\n 1) extract outliers from the data set \n 2)  arrive the data without the outliers")
```

```r
outliers=boxplot(train_data$price,plot=FALSE)$out
outliers_data=train_data[which(train_data$price %in% outliers),]
cat("The Number of Outliers in the actual data is : ")
count(outliers_data)
#excluding the outliers from the data
train_data_excl_ol= train_data[-which(train_data$price %in% outliers),]
```

```r
cat("Plot of original data with and without outliers. ")
par(mfrow=c(1, 2))
plot(train_data$bedrooms, train_data$price, main="Including Outliers", xlab="Bedrooms", ylab="Price", pch="*", col="magenta", cex=2)
abline(lm(price ~ bedrooms, data=train_data), col="green", lwd=3, lty=2)

plot(train_data_excl_ol$bedrooms, train_data_excl_ol$price, main="Exculding Outliers (removed) ", xlab="Bedrooms", ylab="Price", pch="*", col="magenta", cex=2)
abline(lm(price ~bedrooms, data=train_data_excl_ol), col="green", lwd=3, lty=2)

cat("It is important to observe the change of slope.However fitting this without no outliers would be resulting in unrelastic prediction")
```

```r
cat("Here we observe that the change in slope of the best fit line after removing the outliers. It is evident that if we remove the outliers to train the model, our predictions would be exagerated (misleading with high error rate) for larger values of price because of the larger slope.\nAlthough it is the case , we are considering to continue with the entire data for generating the initial model and then fit it to remove outliers ,if we have substantial evidence to remove the same ")
```

```r
cat("Let us begin by generating a Linear model based on the arrived independent variables using lm")

model=lm(data=train_data,price~bedrooms+bathrooms+sqft_living+view+grade+sqft_above+sqft_basement+sqft_living15)
summary(model)
```

```r
par(mfrow=c(2,2))
plot(model)
```

```r
#We can see from the above linear model summary we observe that the relationship between dependant variables and exploratory variables appear to be considerably strong
# which is clearly indicated by the value of  R-Suared value and the probability. Also it can be concluded from the p-value as shown above that sqft_living15 has no significance in the housing price prediction.
# Therfore we could drop it from the model. At the same time we consider any other valid variables which we had ignonred in our previous EDA and arrive at a model which gives us the maximum R-squared 
# value. 
# from the correlation plot we consider the next few variables which had less correlation similar to the exisitng variables in consideration
# sqft_lot , age(~ date - nvl(yr_renovated,nvl(yr_built,0),0 ~ ) , floors , waterfront

model_alter_1=lm(data=train_data,price~bedrooms+bathrooms+sqft_living+view+age+sqft_lot+floors+waterfront)
summary(model_alter_1)
```

```r
par(mfrow=c(2,2))
plot(model_alter_1)
```

```r
cat("From the above model summary it can be concluded that the adjusted R-squared value of 0.601, the relationship beween these variables appear to be quite strong.\nIt could be an imporoper inference to declare an observation as an outlier based on only one (insignificant) feature .\nWhen you have to decide if an individual entity (represented by row or observation) is an extreme value or not, it is better to collectively consider the features (X’s) that matter. We fnd it out using the cook’s distance")
cooksd <- cooks.distance(model_alter_1)
mean(cooksd, na.rm = TRUE)
```

```r
par(mfrow=c(1, 1))
plot(cooksd, main="Influential Observation using Cooks distance",xlim=c(0,25000),ylim=c(0,0.1))
axis(1, at=seq(0, 25000, 5000))
axis(2, at=seq(0, 0.1, 0.0001))
abline(h = 4*mean(cooksd, na.rm=T), col="green")  
text(x=1:length(cooksd)+1,y=cooksd,labels=ifelse(cooksd>4*mean(cooksd,na.rm=T),names(cooksd),""), col="magenta")  
```

```r
cat("From the above data we look to get the influential points affecting the data")
influential <- as.numeric(names(cooksd)[(cooksd > 4*mean(cooksd, na.rm=T))])  
head(train_data[influential, ])
```

```r
influential_data=train_data[influential, ]
```

```r
influencial_outliers=inner_join(outliers_data,influential_data)
```

```r
cat(" We have 13 observations which are outliers yet influential hence we need to keep these outliers.Now we modify the data excluding the outliers and including only the influential outliers.")
train_data_exc_ol_2=rbind(train_data_excl_ol,influencial_outliers)
```

```r
cat("modelling using the train data which includes influential_outliers")
model_alter_2=lm(data=train_data_exc_ol_2,price~bedrooms+bathrooms+sqft_living+view+grade+sqft_lot+age+floors+waterfront)
summary(model_alter_2)
```

```r
par(mfrow=c(2,2))
plot(model_alter_2)
```

```r
anova(model_alter_2)
```

```r
cat("We can see the relationship beween these variables appear to be moderately strong as shown by R-Suared value and the probability.also coclude from the p-value that sqft_lot is not a significant variable for the prediction of price.\nHence we drop it. We also try fitting the model including a few other variables which we left out in while fitting the model when we took the entire data and stop at a model which gives us the maximum R-squared value.")
model_alter_3=lm(data=train_data_exc_ol_2,price~bedrooms+bathrooms+sqft_living+view+grade+age+waterfront+long+lat+zipcode+condition+sqft_above+sqft_living15+reno)
summary(model_alter_3)
```

```r
cat("As concluded from the adjusted R-squared value of 0.6907 , the relationship beween these variables appear to be quite strong.")
```

```r
par(mfrow=c(2,2))
plot(model_alter_3)
```

```r
pred=model_alter_3$fitted.values

tally_table=data.frame(actual=train_data_exc_ol_2$price, predicted=pred)
head(tally_table)
mape=mean(abs(tally_table$actual-tally_table$predicted)/tally_table$actual)
accuracy=1-mape
cat("theTraining accuracy arrived for the model is:", round(accuracy*100,2),"%.")
```

```r
date_sale_test=substring(test_data$date, first=1, last=4)

test_data$date_sale_test = date_sale_test
test_data$sale_date_year=as.integer(date_sale_test)
test_data$age=test_data$sale_date_year-test_data$yr_built

test_data$reno=ifelse(test_data$yr_renovated==0,0,1)
test_data$reno=as.factor(test_data$reno)

#bedrooms+bathrooms+sqft_living+view+grade+age+waterfront+long+lat+zipcode+condition+sqft_above+sqft_living15+reno

test_data_1=test_data[,c(4,5,6,10,12,24,9,19,18,17,11,13,20,25)]

pred_test=predict(newdata=test_data,model_alter_3)
```

```r
arrived_table=data.frame(actual=test_data$price, predicted=pred_test)

obsrv_test_mean=mean(abs(arrived_table$actual-arrived_table$predicted)/arrived_table$actual)
test_accuracy_score=1-obsrv_test_mean

cat("The accuracy arrived using the test data for the model is:", test_accuracy_score)
```

```r
cat("Hence it can be interpreted that the accuracy is :",round(test_accuracy_score*100,2),"%.\nThus , the arrived model can perform thehousing price prediciton with an accuracy of ",round(test_accuracy_score*100,2),"%.")
```
